## Inspection + make lists

Importera och välj events-fil

Inspektera filen och se vilka unika keys som finns i den

```python
import uproot
import matplotlib.pyplot as plt
import awkward as ak

events_path = "../src/mldmx/data/overlay_main10_pileup20_00/pileup.root"
figures_path = "../figures/notebooks/"

file = uproot.open(events_path)
tree = file["LDMX_Events"]

f_inspect = uproot.open(events_path)

keys = [k.split(";")[0] for k in f_inspect.keys()]  # strip ROOT cycle suffix ";1"
keys = sorted(set(keys))

print(f"Found {len(keys)} unique objects in {events_path}\n")
for k in keys:
    print(k)

print("\nBRANCHES (no leaves):")
branches_no_leaves = sorted({b.split("/")[0] for b in tree.keys()})
for b in branches_no_leaves:
    print(b)

print("\nBRANCHES:")
for b in tree.keys():
    print(b)
```

Number of events, overlay tree true/false tag

```python
num_events = tree.num_entries
print(f"\nNumber of events in the tree: {num_events}")

# Overlay tree tag
overlay_tree = any("overlay" in b for b in tree.keys())

if overlay_tree:
    print("\nThis is an overlay tree file")
else:
    print("\nThis is not an overlay tree file")
```

## Plots!

Energy distribution, 2D plots, 3D plot

Hitta EcalRecHits i x, y, z

Hitta energierna för de olika hits som finns och plotta i histogram

```python
# To array x, y ,z and energy of Ecal RecHits
if overlay_tree:
    # x,y,z and reconstructed energy of ECal 
    ecal_y = tree["EcalRecHits_overlay/EcalRecHits_overlay.ypos_"].array()
    ecal_x = tree["EcalRecHits_overlay/EcalRecHits_overlay.xpos_"].array()
    ecal_z = tree["EcalRecHits_overlay/EcalRecHits_overlay.zpos_"].array()
    ecal_energy = tree["EcalRecHits_overlay/EcalRecHits_overlay.energy_"].array()
else: # x,y,z and **deposited** energy in ECal
    ecal_y = tree["EcalSimHits_pileup/EcalSimHits_pileup.y_"].array()
    ecal_x = tree["EcalSimHits_pileup/EcalSimHits_pileup.x_"].array()
    ecal_z = tree["EcalSimHits_pileup/EcalSimHits_pileup.z_"].array()
    ecal_energy = tree["EcalSimHits_pileup/EcalSimHits_pileup.edep_"].array()
    # Energy of particle when crossing detector surfaces (truth kinematics)
    # TargetScoringPlaneHits_pileup/TargetScoringPlaneHits_pileup.energy_
    
    # SimHit deposited energies (truth detector response)
    # EcalSimHits_pileup/EcalSimHits_pileup.edep_

    # Particle (truth) energy
    # SimParticles_pileup/SimParticles_pileup.second.energy_

# Create figure
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
ax.hist(ak.flatten(ecal_energy), bins=100, alpha=0.7, color='blue', range=[0,400])
if overlay_tree:
    ax.set_title("Ecal RecHit Energy Distribution")
else:
    ax.set_title("Ecal SimHit Energy Distribution")
ax.set_xlabel("Energy (MeV)")
ax.set_ylabel("Counts")
plt.savefig(f"{figures_path}ecal_energy.png", dpi=300)
#plt.close()
```

### 2D: z vs x plot

```python
# Scatter plots of x vs z for up to first 9 events (3x3 grid)

rows, columns = 3, 3
nplots = rows * columns
nevents = len(ecal_x)
nshow = min(nplots, nevents)

fig, ax = plt.subplots(rows, columns, figsize=(12, 12))
ax = ax.flatten()

for index in range(nshow):
    ax[index].scatter(ecal_x[index], ecal_z[index], s=4)
    ax[index].set_title(f"Event {index}")
    ax[index].set_xlabel("X Position [mm]")
    ax[index].set_ylabel("Z Position [mm]")
    ax[index].set_xlim(-300, 300)
    ax[index].set_ylim(200, 700)

# turn off unused panels if fewer than 9 events
for index in range(nshow, nplots):
    ax[index].axis("off")

plt.tight_layout()

plt.savefig(f"{figures_path}ecal_x_z_hits_3x3.png", dpi=300)
plt.show()
```

### 2D: y vs z plot

```python
# Scatter plots of y vs z for up to first 9 events (3x3 grid)

rows, columns = 3, 3
nplots = rows * columns
nevents = len(ecal_y)
nshow = min(nplots, nevents)

fig, ax = plt.subplots(rows, columns, figsize=(12, 12))
ax = ax.flatten()

for index in range(nshow):
    ax[index].scatter(ecal_y[index], ecal_z[index], s=4)
    ax[index].set_title(f"Event {index}")
    ax[index].set_xlabel("Y Position [mm]")
    ax[index].set_ylabel("Z Position [mm]")
    ax[index].set_xlim(-300, 300)
    ax[index].set_ylim(200, 700)

# turn off unused panels
for index in range(nshow, nplots):
    ax[index].axis("off")

plt.tight_layout()
plt.savefig(f"{figures_path}ecal_y_z_hits_3x3.png", dpi=300)
plt.show()
```

### 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D projection
import awkward as ak
import matplotlib.pyplot as plt

event_idx = 0

# convert one event to regular numpy arrays
x = ak.to_numpy(ecal_x[event_idx])
y = ak.to_numpy(ecal_y[event_idx])
z = ak.to_numpy(ecal_z[event_idx])
e = ak.to_numpy(ecal_energy[event_idx])

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")

sc = ax.scatter(x, y, z, c=e, s=10, alpha=0.8)

ax.set_title(f"ECal RecHits in 3D, Event {event_idx}")
ax.set_xlabel("X Position [mm]")
ax.set_ylabel("Y Position [mm]")
ax.set_zlabel("Z Position [mm]")

ax.set_xlim(-300, 300)
ax.set_ylim(-300, 300)
ax.set_zlim(200, 700)

cbar = plt.colorbar(sc, ax=ax, pad=0.1)
cbar.set_label("Energy [MeV]")

plt.tight_layout()
plt.savefig(f"{figures_path}ecal_hits_3d_event0.png", dpi=300)
plt.show()
```

### Interactive 3D

```python
import plotly.graph_objects as go

event_idx = 0

x = ak.to_numpy(ecal_x[event_idx])
y = ak.to_numpy(ecal_y[event_idx])
z = ak.to_numpy(ecal_z[event_idx])
e = ak.to_numpy(ecal_energy[event_idx])

fig = go.Figure(
    data=[
        go.Scatter3d(
            x=x,
            y=y,
            z=z,
            mode="markers",
            marker=dict(
                size=3,
                color=e,
                colorscale="Viridis",
                colorbar=dict(title="Energy [MeV]"),
                opacity=0.8,
            ),
        )
    ]
)

fig.update_layout(
    title=f"ECal {('RecHits' if overlay_tree else 'SimHits')} 3D (Event {event_idx})",
    scene=dict(
        xaxis_title="X [mm]",
        yaxis_title="Y [mm]",
        zaxis_title="Z [mm]",
        xaxis=dict(range=[-300, 300]),
        yaxis=dict(range=[-300, 300]),
        zaxis=dict(range=[200, 700]),
    ),
    margin=dict(l=0, r=0, b=0, t=40),
)

fig.show()
```
